```python
from triangulator.triangulator import triangulate
```

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib.patches import Polygon
```

```python
polygon = ((1, 1), (4, 2), (3, 2), (2, 7), (3, 4), (7, 3), (8, 3), (9, 0))
```

```python
drawable_polygon = Polygon(polygon)
```

```python
plt.plot(*zip(*drawable_polygon.get_xy()), color='red')
```

```python
triangles =  triangulate(polygon)
```

```python
for triangle in triangles:
    edge1 = (triangle[0], triangle[1])
    edge2 = (triangle[1], triangle[2])
    edge3 = (triangle[2], triangle[0])
    plt.plot(*zip(*edge1), color='red')
    plt.plot(*zip(*edge2), color='red')
    plt.plot(*zip(*edge3), color='red')
```
